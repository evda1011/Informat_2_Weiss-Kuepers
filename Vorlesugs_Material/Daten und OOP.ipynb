### Beispiel: Listen 
- Vgl. https://docs.python.org/3/tutorial/datastructures.html
- und https://docs.python.org/3/library/stdtypes.html#list

```python
# Exemplary list operations
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits.count('apple'))
print(fruits.count('tangerine'))
# print(fruits.index('banana‘))
# find next banana starting a position 4
print(fruits.index('banana', 4))
fruits.reverse()
print(fruits)

fruits.append('grape')
print(fruits)

fruits.sort()
print(fruits)

print(fruits.pop())
print(fruits)
```

### Beispiel: Dictionaries
- Vgl. https://docs.python.org/3/library/stdtypes.html#dict

```python
# dictionary iteration
capitols = {'Germany': 'Berlin', 'France': 'Paris'}
for k, v in capitols.items():
    print(k, v)

# access only the keys or values
print(capitols.keys())
print(capitols.values())

# update / add new dictionary to existing one
capitols.update({'Germany':'Berlin', 'Ireland':'Dublin'})
print(capitols)

# "in"-notation to check existence of keys
print("Ireland" in capitols.keys())

# pop-method in analogy to its usage with lists
print(capitols.pop('France'))
print(capitols)
```

- Beispiel: „Pandas DataFrame“

- Wir können eine tabellarische Datenstruktur in einem pandas
DataFrame speichern.
    - Dazu wird ein entsprechendes Objekt erzeugt, wenn wir den Konstruktor richtig nutzen:
    -  Wenn wir uns mit der API von pandas DataFrames auseinandersetzen, können wir bspw. Analysen in Analogie zu „SQL Group By“ durchführen:



```python
# create example tabular data structure (pandas DataFrame)
import pandas as pd

df_sales = pd.DataFrame([['A', 10], ['B', 20], ['B', 15], ['A', 5]], columns=['product', 'sales'])
print(df_sales)

# use exemplary analysis-method on the DataFrame object
print(df_sales.groupby('product').agg({'sales' : ['mean', 'min', 'max']}))

# load dataset 'mpg' into dataframe
import seaborn as sns

df_mpg = sns.load_dataset('mpg')
print(df_mpg.groupby(['origin']).agg({'mpg':'mean', 'horsepower':'mean'}))
```

## Iteratoren

- Die Objekte, die wir mittels for-Schleife auf diese Weise „durchgehen“ nennen sich iterierbar („iterable“).
    - Man kann einen Iterator von einem iterierbaren Objekt erzeugen.
        - Dies passiert implizit bei der Verwendung einer for-Schleife.
    - Wir können es auch explizit umsetzen, aber for-Schleifen sind eleganter

```python
# explizite Erstellung und Verwendung
# eines Iterator-Objekts
x = iter(fruits)
print(x)
print(next(x))
print(next(x))
print(next(x)) # kein Objekt mehr! --> Exception
```

```python
# explizite Erstellung und Verwendung
# eines Iterator-Objekts
x = iter(fruits)
print(x)
print(next(x))
print(next(x))
print(next(x)) # kein Objekt mehr! --> Exception

# so könnte die vollständige
# Iteration umgesetzt werden,
# aber for-loops sind "eleganter"
x = iter(fruits)
try:
    while True:
        print(next(x))
except StopIteration:
    pass
```
