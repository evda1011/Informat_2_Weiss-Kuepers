### Welche arten von elementen in python gibt es

In Python gibt es verschiedene Arten von Elementen (Datenobjekten), die man nach ihrer grundlegenden Natur kategorisieren kann. Diese sind die Basisdatentypen, Datenstrukturen und Objekttypen, die in Python verwendet werden. Nachfolgend eine Übersicht:
1. Primitive Datentypen
- Integer (int): Ganze Zahlen, z. B. 42, -7
- Float (float): Gleitkommazahlen, z. B. 3.14, -0.001
- Boolean (bool): Wahrheitswerte, True oder False
- String (str): Zeichenketten, z. B. "Hallo Welt", 'Python'
- NoneType (None): Repräsentiert das Fehlen eines Wertes, None
2. Sequenzen
- List (list): Veränderbare Sequenzen von beliebigen Elementen, z. B. [1, 2, 3, "Hallo"]
- Tuple (tuple): Unveränderbare Sequenzen, z. B. (1, 2, 3)
- Range (range): Eine Folge von Zahlen, oft in Schleifen verwendet, z. B. range(0, 5)
3. Sets und Mappings
- Set (set): Menge von eindeutigen, unveränderbaren Elementen, z. B. {1, 2, 3}
- Frozenset (frozenset): Unveränderliche Mengen
- Dictionary (dict): Schlüssel-Wert-Paare, z. B. {"Name": "Anna", "Alter": 25}
4. Weitere Objekte
- Bytes (bytes): Unveränderbare Sequenzen von Bytes, z. B. b'Hallo'
- Bytearray (bytearray): Veränderbare Sequenzen von Bytes
- Complex (complex): Komplexe Zahlen, z. B. 3 + 4j
5. Benutzerdefinierte Typen
In Python kann man eigene Klassen (class) definieren, um komplexe Objekte zu erstellen. Objekte dieser Klassen sind Instanzen und können beliebige Attribute und Methoden besitzen.
6. Funktionale Elemente
- Funktionen (def bzw. lambda): Können wie Objekte behandelt und gespeichert werden
- Module und Pakete: Sammlungen von Funktionen, Klassen und Variablen
- Beispiel für verschiedene Datentypen


Zusammenfassend lässt sich sagen, dass Python eine reichhaltige Typensystemstruktur bietet, von einfachen primitiven Typen bis hin zu komplexen Datenstrukturen und benutzerdefinierten Klassen. Diese Flexibilität macht Python besonders für schnelle Entwicklung und Datenverarbeitung geeignet.

```python
# Primitive Datentypen
x = 5           # int
pi = 3.14       # float
flag = True     # bool
name = "Max"    # str
nothing = None  # NoneType

# Sequenzen
numbers = [1, 2, 3, 4]           # list
coords = (10, 20)                 # tuple
nums_range = range(5)             # range

# Mengen und Dictionaries
unique_set = {1, 2, 3}            # set
info = {"Name": "Anna", "Alter": 25}  # dict

# Bytes und komplexe Zahlen
b = b"Hallo"                       # bytes
c = 1 + 2j                         # complex
```

### Was ist das Konzept mutable & immutable
#### 1. Mutable (veränderlich)
Definition: Der Inhalt eines Objekts kann nach der Erstellung geändert werden.

Änderungen erfolgen in-place (am selben Speicherort).

Beispiele:
- list
- dict
- set
- Benutzerdefinierte Klassen (standardmäßig)

```python
print("Mutable")
liste = [1, 2, 3]
print(id(liste))
liste.append(4)
print(id(liste))
```

```python
numbers = [1, 2, 3]
print(id(numbers))  # Speicheradresse vor Änderung
numbers.append(4)   # Liste wird verändert
print(numbers)      # [1, 2, 3, 4]
print(id(numbers))  # gleiche Speicheradresse
```

#### 2. Immutable (unveränderlich)

Definition: Der Inhalt eines Objekts kann nicht geändert werden.
Jede "Änderung" erzeugt ein neues Objekt mit neuer Speicheradresse.

Beispiele:
- int
- float
- str
- tuple
- frozenset

```python

text = "Hallo"
print(id(text))     # Speicheradresse vor Änderung
text += " Welt"     # erzeugt ein neues String-Objekt
print(text)         # "Hallo Welt"
print(id(text))     # neue Speicheradresse
```
